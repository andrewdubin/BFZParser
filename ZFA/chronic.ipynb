```python
import pandas as pd
import numpy as np
```

```python
#Step 1: Import document, Change Program Start Date to Date of Identification
df = pd.read_csv("June_Main.csv",parse_dates=['Program Start Date','Program End Date'])
df = df.rename(columns={"Program Start Date":"Date of Identification",'Case Number':'Client ID','Veteran Status (HUD)':'Veteran Status'})
df = df.drop(['Name','Family Full SPDAT Total:_2052','Total Single SPDAT:_2167','Gender (HUD)','Race (HUD)','Ethnicity (HUD)',
              'Participant Site Identifier','Program Unique Identifier'],axis=1)
df = df.drop_duplicates()
df = df.reset_index(drop=True)
df
```

```python
chronic = pd.read_csv("June_Chronic.csv",usecols=['Case Number','CH/Not CH'])
chronic = chronic.rename(columns={"Case Number":'Client ID'})
chronic = chronic[chronic['CH/Not CH']=='CH']
chronic['CH/Not CH'] = "Yes"
chronic = chronic.drop_duplicates()
chronic
```

```python
df.merge(chronic,on='Client ID',how='left')['CH/Not CH'].value_counts()
```

```python
df['Chronic Alt'] = df['Client ID'].map(dict(zip(chronic['Client ID'],chronic['CH/Not CH'])))
df[df['Chronic Alt']=='Yes']
```
